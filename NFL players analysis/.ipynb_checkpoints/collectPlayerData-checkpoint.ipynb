```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime

# Import API key
from api_keys import sport_key
```

```python
# Take in all of our wrestling data and read it into pandas
stds_2014 = "Resources/standings_2014.csv"
stds_2015 = "Resources/standings_2015.csv"
stds_2016 = "Resources/standings_2016.csv"
stds_2017 = "Resources/standings_2017.csv"
stds_2018 = "Resources/standings_2018.csv"

stds_2014_df = pd.read_csv(stds_2014)
stds_2015_df = pd.read_csv(stds_2015)
stds_2016_df = pd.read_csv(stds_2016)
stds_2017_df = pd.read_csv(stds_2017)
stds_2018_df = pd.read_csv(stds_2018)
```

```python
# Append all datasets
combined_standings_df = stds_2014_df
combined_standings_df = combined_standings_df.append(stds_2015_df)
combined_standings_df = combined_standings_df.append(stds_2016_df)
combined_standings_df = combined_standings_df.append(stds_2017_df)
combined_standings_df = combined_standings_df.append(stds_2018_df)

combined_standings_df.to_csv(f"Resources/combined_standings.csv", index=False)

combined_standings_df
```

```python
# Take in all of our wrestling data and read it into pandas
all_players_2014 = "Resources/all_players_2014.csv"
all_players_2015 = "Resources/all_players_2015.csv"
all_players_2016 = "Resources/all_players_2016.csv"
all_players_2017 = "Resources/all_players_2017.csv"
all_players_2018 = "Resources/all_players_2018.csv"

all_players_2014_df = pd.read_csv(all_players_2014)
all_players_2015_df = pd.read_csv(all_players_2015)
all_players_2016_df = pd.read_csv(all_players_2016)
all_players_2017_df = pd.read_csv(all_players_2017)
all_players_2018_df = pd.read_csv(all_players_2018)
```

```python
# Append all datasets
combined_playing_years_df = all_players_2014_df
combined_playing_years_df = combined_playing_years_df.append(all_players_2015_df)
combined_playing_years_df = combined_playing_years_df.append(all_players_2016_df)
combined_playing_years_df = combined_playing_years_df.append(all_players_2017_df)
combined_playing_years_df = combined_playing_years_df.append(all_players_2018_df)

combined_playing_years_df.to_csv(f"Resources/combined_playing_years_df.csv", index=False)

combined_playing_years_df.head()
```

```python
all_QB = combined_playing_years_df[combined_playing_years_df['Position']=='QB']
all_QB = all_QB.reset_index()
all_QB = all_QB.drop(columns = ['index'])
all_QB.head()
```

```python
players_bio = []

players_bio.append({
    'player_id'   : player_json['id'],
    'player_name' : player_json['name'],
    'jersey_no'   : player_json['jersey'],
    'birth_date'  : player_json['birth_date'], 
    'weight'      : player_json['weight'],
    'height'      : player_json['height'],
    'position'    : player_json['position'],
    'birth_place' : player_json['birth_place'],
    'high_school' : player_json['high_school'], 
    'college'     : player_json['college'],
    'college_conf': player_json['college_conf'], 
    'rookie_year' : player_json['rookie_year'], 
    'status'      : player_json['status'],
    '_timestamp'  : round(time.time())
})

players_bio_df = pd.DataFrame(players_bio)
players_bio_df.to_csv(f"Resources/players_bio.csv", index=False)
players_bio_df.head()
```

```python
players_draft = []

players_draft.append({          
    'player_id'   : player_json['id'],
    'player_name' : player_json['name'],
    'position'    : player_json['position'],
    
    'draft_year'  : player_json['draft']['year'],
    'draft_round' : player_json['draft']['round'], 
    'draft_number': player_json['draft']['number'],
    'team_id'     : player_json['draft']['team']['id'],
    'team_name'   : player_json['draft']['team']['name'],
    'team_market' : player_json['draft']['team']['market'],
    
    '_timestamp'  : round(time.time())
})

players_draft_df = pd.DataFrame(players_draft)
players_draft_df.to_csv(f"Resources/players_draft.csv", index=False)
players_draft_df.head()
```

```python
players_seasons = []

for season in range(len(player_json['seasons'])):
    players_seasons.append({          
        'player_id'   : player_json['id'],
        'player_name' : player_json['name'],
        'position'    : player_json['position'],
        
        'season_year'  : player_json['seasons'][season]['year'],
        'season_type'  : player_json['seasons'][season]['type'], 
        'team_id'      : player_json['seasons'][season]['teams'][0]['id'],
        'team_name'    : player_json['seasons'][season]['teams'][0]['name'],
        'team_market'  : player_json['seasons'][season]['teams'][0]['market'],
        'games_played' : player_json['seasons'][season]['teams'][0]['statistics']['games_played'],
        'games_started': player_json['seasons'][season]['teams'][0]['statistics']['games_started'],
        
        '_timestamp'  : round(time.time())
    })

players_seasons_df = pd.DataFrame(players_seasons)
players_seasons_df.to_csv(f"Resources/players_seasons.csv", index=False)
players_seasons_df.head()
```

```python
player_stats_rushing = []

for season in range(len(player_json['seasons'])):
    if player_json['seasons'][season]['teams'][0]['statistics']['games_played'] > 0:
        if 'rushing' in player_json['seasons'][season]['teams'][0]['statistics'].keys():
        
            player_stats_rushing.append({          
                'player_id'   : player_json['id'],
                'player_name' : player_json['name'],
                'position'    : player_json['position'],

                'season_year'  : player_json['seasons'][season]['year'],
                'season_type'  : player_json['seasons'][season]['type'], 
                'team_id'      : player_json['seasons'][season]['teams'][0]['id'],
                'team_name'    : player_json['seasons'][season]['teams'][0]['name'],
                'team_market'  : player_json['seasons'][season]['teams'][0]['market'],

                'avg_yards'          : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['avg_yards'],
                'attempts'           : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['attempts'],
                'touchdowns'         : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['touchdowns'],
                'yards'              : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['yards'],
                'longest'            : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['longest'],
                'longest_touchdown'  : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['longest_touchdown'],        
                'redzone_attempts'   : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['redzone_attempts'],
                'tlost'              : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['tlost'],
                'tlost_yards'        : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['tlost_yards'],
                'broken_tackles'     : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['broken_tackles'],
                'kneel_downs'        : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['kneel_downs'],
                'scrambles'          : player_json['seasons'][season]['teams'][0]['statistics']['rushing']['scrambles'],
                'yards_after_contact': player_json['seasons'][season]['teams'][0]['statistics']['rushing']['yards_after_contact'],             

                '_timestamp'  : round(time.time())
        })

player_stats_rushing_df = pd.DataFrame(player_stats_rushing)
player_stats_rushing_df.to_csv(f"Resources/player_stats_rushing.csv", index=False)
player_stats_rushing_df.head()
```
