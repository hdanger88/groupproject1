```python
data_year = '2014'
standings_url = f"http://api.sportradar.us/nfl/official/trial/v5/en/seasons/{data_year}/standings.json?"

standings_query_url = f"{standings_url}api_key={sport_key}"

standings_response = requests.get(standings_query_url)
standings_json = standings_response.json()
```

```python
teams = []
team_data = []

for conference in range(len(standings_json['conferences'])):

    for division in range(len(standings_json['conferences'][conference])):
        
        for team in range(len(standings_json['conferences'][conference]['divisions'][division])):
            
            teams.append({
            'Conference'    : standings_json['conferences'][conference]['name'],
            'Division'      : standings_json['conferences'][conference]['divisions'][division]['name'],
            'Team_Name'     : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['name'],
            'Team_ID'       : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['id'],
            'Wins'          : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['wins'],
            'Losses'        : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['losses'],         
            'Ties'          : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['ties'],
            'Win_pct'       : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['win_pct'],
            'Points_for'    : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['points_for'],
            'Points_against': standings_json['conferences'][conference]['divisions'][division]['teams'][team]['points_against'],
            'Points_rank'   : standings_json['conferences'][conference]['divisions'][division]['teams'][team]['points_rank'],
            'Touchdown_diff': standings_json['conferences'][conference]['divisions'][division]['teams'][team]['touchdown_diff'],
        })
            
teams_df = pd.DataFrame(teams)
teams_df.groupby(['Conference','Division', 'Team_Name']).mean()
teams_df.to_csv(f"Output/standings_{data_year}.csv")
```
