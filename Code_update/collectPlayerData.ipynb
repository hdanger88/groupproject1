```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime

# Import API key
from api_keys import sport_key
```

```python
# Take in all of our wrestling data and read it into pandas
stds_2014 = "Resources/standings_2014.csv"
stds_2015 = "Resources/standings_2015.csv"
stds_2016 = "Resources/standings_2016.csv"
stds_2017 = "Resources/standings_2017.csv"
stds_2018 = "Resources/standings_2018.csv"

stds_2014_df = pd.read_csv(stds_2014)
stds_2015_df = pd.read_csv(stds_2015)
stds_2016_df = pd.read_csv(stds_2016)
stds_2017_df = pd.read_csv(stds_2017)
stds_2018_df = pd.read_csv(stds_2018)
```

```python
# Append all datasets
combined_standings_df = stds_2014_df
combined_standings_df = combined_standings_df.append(stds_2015_df)
combined_standings_df = combined_standings_df.append(stds_2016_df)
combined_standings_df = combined_standings_df.append(stds_2017_df)
combined_standings_df = combined_standings_df.append(stds_2018_df)

combined_standings_df.to_csv(f"Resources/combined_standings.csv", index=False)

combined_standings_df
```
